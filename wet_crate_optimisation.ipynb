```python
def wasted(so_quantity, size):
    if size==0:
        return 100
    a=so_quantity%size
    if a==0:
        return a
    else:
        return size-a
```

```python
def crates_used(so_quantity, size):
    if size==0:
        return 100
    if size==0:
        return 0
    return math.ceil(so_quantity/size)
```

```python
master["Small_wasted_volume"] = master.apply(lambda row: wasted(row["SO Quantity"], row["Small"]), axis=1)
master["Medium_wasted_volume"] = master.apply(lambda row: wasted(row["SO Quantity"], row["Medium"] ), axis=1)
master["Large_wasted_volume"] = master.apply(lambda row: wasted(row["SO Quantity"], row["Large"]), axis=1)
```

```python
master["Small_crates_used"] = master.apply(lambda row: crates_used(row["SO Quantity"], row["Small"]), axis=1)
master["Medium_crates_used"] = master.apply(lambda row: crates_used(row["SO Quantity"], row["Medium"] ), axis=1)
master["Large_crates_used"] = master.apply(lambda row: crates_used(row["SO Quantity"], row["Large"]), axis=1)

master["old_crates_used"] = master.apply(lambda row: crates_used(row["SO Quantity"], row["Old crate"]), axis=1)
```

```python
master["Small_crates_used"] = master["Small_crates_used"] - master["old_crates_used"]
master["Medium_crates_used"] = master["Medium_crates_used"] - master["old_crates_used"]
master["Large_crates_used"] = master["Large_crates_used"] - master["old_crates_used"]
```

```python
# Assuming you have a DataFrame called "df" with a column "Material Description" and columns for "Large Crates", "Medium Crates", and "Small Crates"

# Group the data by "Material Description" and calculate the sum of each crate type
crate_sums = master.groupby("Material Description")["Small_wasted_volume", "Medium_wasted_volume", "Large_wasted_volume","Small_crates_used","Medium_crates_used","Large_crates_used","old_crates_used"].sum()
crate_sums = crate_sums.reset_index()
crate_sums
```

```python
def w2c(wasted_volume, crate_values):
    output = wasted_volume.copy()  # Create a copy of the series to store the output values
    
    for i in range(len(wasted_volume)):
        if crate_values.iloc[i] == 0:
            output.iloc[i] = wasted_volume.iloc[i]
        else:
            output.iloc[i] = wasted_volume.iloc[i] / crate_values.iloc[i]
    
    return output
```
